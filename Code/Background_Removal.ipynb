# Background Removal for Train data

```python
%cd '/content/gdrive/MyDrive/background_removal_DL'
!python train_u2net_image.py
```

```python
import glob
import os
result_dir='/content/gdrive/MyDrive/background_removal_DL/test_data/images/train_u2net_results'
result_img_name_list = glob.glob(result_dir + os.sep + '*')
#orginalimage (CHANGE FILE EXTENSION HERE - BY DEFAULT: .jpg)
original='/content/gdrive/MyDrive/VideoFrames/TrainingFrames'
original_img_name_list = glob.glob(original + os.sep + '*')
print(len(result_img_name_list))

print(original_img_name_list)
result_img_name_list[406].split(os.sep)[-1] 
```

```python
len(result_img_name_list), len(original_img_name_list)
```

```python
#subimage
import cv2
from PIL import Image
for i in range(len(result_img_name_list)):
  u2netresults=cv2.imread(result_img_name_list[i])
  original=cv2.imread(original_img_name_list[i])
  subimage=cv2.subtract(u2netresults,original)
  cv2.imwrite('/content/gdrive/MyDrive/VideoFrames/bg removal/bgTrainingFrames/'+result_img_name_list[i].split(os.sep)[-1],subimage)
  subimage = Image.open('/content/gdrive/MyDrive/VideoFrames/bg removal/bgTrainingFrames/'+result_img_name_list[i].split(os.sep)[-1])
  original = Image.open(original_img_name_list[i])
  subimage=subimage.convert("RGBA")
  original=original.convert("RGBA")
  subdata=subimage.getdata()
  ogdata=original.getdata()
  newdata=[]
  for j in range(subdata.size[0]*subdata.size[1]):
    if subdata[j][0]==0 and subdata[j][1]==0 and subdata[j][2]==0:
      newdata.append((255,255,255,0))
    else:
      newdata.append(ogdata[j])
  subimage.putdata(newdata)
  #cv2.imwrite('/content/gdrive/MyDrive/bg remove/testt/'+result_img_name_list[i].split(os.sep)[-1],subimage)
  subimage.save('/content/gdrive/MyDrive/VideoFrames/bg removal/bgTrainingFrames/'+result_img_name_list[i].split(os.sep)[-1])
```

```python
import cv2
import os
```

```python
def getFrame(sec, vidcap, count):
    vidcap.set(cv2.CAP_PROP_POS_MSEC,sec*1000)
    hasFrames, image = vidcap.read()

    if hasFrames:
        filename = '/content/gdrive/MyDrive/VideoFrames/Evaluate/' + "%d.png" % count
        cv2.imwrite(filename, image)    # save frame as JPG file
        print(filename)

    return hasFrames
```

```python
def createFrames(video):
  
  path = '/content/gdrive/MyDrive/VideoFrames/Evaluate'
  files = os.listdir(path)
  print('removing files')
  for f in files:
    print(f)
    os.remove(path+'/'+f)

  print('creating frames')
  vidcap = cv2.VideoCapture(video)
  sec = 0
  frameRate = 1/6
  count = 1 
  success = getFrame(sec, vidcap, count)

  while success:
    count = count + 1
    sec = sec + frameRate
    sec = round(sec, 2)
    success = getFrame(sec, vidcap, count)
```

```python
createFrames('/content/gdrive/MyDrive/PSL Schools Dataset/front/Moiz/sentence9.mp4')
```

```python
#for 90 degree clockwise rotation
path = '/content/gdrive/MyDrive/VideoFrames/Evaluate'
files = os.listdir(path)
for f in files:
  src = cv2.imread(path+'/'+f)
  
  # Using cv2.rotate() method
  # Using cv2.ROTATE_90_CLOCKWISE rotate
  # by 90 degrees clockwise
  image = cv2.rotate(src, cv2.cv2.ROTATE_90_CLOCKWISE)
  cv2.imwrite(path+'/'+f, image)
```
