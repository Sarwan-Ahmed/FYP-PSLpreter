create txt file of videos paths

```python
import os

text_file = open("videosfront.txt", "w")
path = "/content/drive/MyDrive/PSL Schools Dataset/front"
dirs = os.listdir(path)

for person in os.listdir(path):
  for video in os.listdir(path+"/"+person):
    video_name = path+"/"+person+"/"+video
    text_file.write(video_name+"\n")
    print(video_name)

text_file.close()
```

Create Training frames

```python
import cv2

def getFrame(sec, frame_name, vidcap, count):
    vidcap.set(cv2.CAP_PROP_POS_MSEC,sec*1000)
    hasFrames, image = vidcap.read()

    if hasFrames:
        #image = cv2.resize(image, dsize=(120,120))              #resize image to 120x120
        filename = '/content/drive/MyDrive/VideoFrames/TrainingFrames/' + frame_name + "_%d.jpg" % count
        cv2.imwrite(filename, image)    # save frame as JPG file

    return hasFrames
```

```python
# open the .txt file which have names of training videos
f = open("/content/drive/MyDrive/PSL dataset/trainingVideosFront2.txt", "r")
temp = f.read()
videos = temp.split('\n')


# storing the frames from training videos
for video in videos:

    vidcap = cv2.VideoCapture(video)
    sec = 0
    frameRate = 1/6  # it will capture image in each 1/6th second
    count = 1
    #frame_name = video.replace('/', '_')
    video_array = video.split("/")
    print(video)

    frame_name = video_array[-3] + "_" + video_array[-2] + "_" + video_array[-1][:-4]
    print(frame_name+"\n")
    
    success = getFrame(sec, frame_name, vidcap, count)

    while success:
        count = count + 1
        sec = sec + frameRate
        sec = round(sec, 2)
        success = getFrame(sec, frame_name, vidcap, count)
```
